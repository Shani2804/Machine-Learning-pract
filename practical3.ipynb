```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dat['NumOfProduct'].map({1:'one',2:'two',3:'three',4:'four'})
```

```python
dat=pd.get_dummies(dat,columns=['NumOfProduct'])
dat.head()
```

```python
x=dat.drop(["CustomerId","Surname","Exited"],axis=1)
y=dat["Exited"]
x= pd.get_dummies(x, columns=["Geography", "Gender"])
```

```python
datage=dat.groupby(['Age']).count()
datage.tail(10)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
corr_mat=dat.select_dtypes(include=np.number).corr()
sns.heatmap(corr_mat,vmax=1,vmin=-1,annot=True,linewidths=.5,cmap='coolwarm')
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
y_test
```

```python
from sklearn.metrics import mean_absolute_percentage_error as mape
mape_value=mape(y_test,y_pred)
mape_value
```

```python
from sklearn.metrics import mean_squared_error as mse
mse_value=mse(y_test,y_pred)
mse_value
```

```python
from sklearn.metrics import mean_absolute_error as mae
mae_value=mae(y_test,y_pred)
mae_value
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
acc
```
